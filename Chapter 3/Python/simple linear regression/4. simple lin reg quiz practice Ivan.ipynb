```python
from statsmodels.formula.api import ols
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [400, 450, 500, 520, 600]
y = [480, 500, 550, 600, 750]
df = pd.DataFrame({"x": x, "y": y})
df
```

```python
plt.scatter(x, y)
```

```python
reg = ols(formula="y ~ x", data=df)
result = reg.fit()
print(result.summary())
```

```python
xx = np.linspace(min(x), max(x), 50)
y_pred = result.predict(pd.DataFrame({"x": xx}))
plt.plot(xx, y_pred)
plt.scatter(x, y)
```
