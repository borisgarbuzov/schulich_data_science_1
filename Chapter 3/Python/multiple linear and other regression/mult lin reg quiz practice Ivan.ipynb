```python
from statsmodels.formula.api import ols
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x = [400, 450, 500, 520, 600]
# y = [480, 500, 550, 600, 750]
x = [1000, 780, 590, 470, 500, 600]
y = [1500, 800, 700, 600, 660, 780]
df = pd.DataFrame({"x": x, "y": y})
df
```

```python
plt.scatter(x, y)
```

```python
reg = ols(formula="y ~ x", data=df)
result = reg.fit()
result.summary()
```

```python
xx = np.linspace(min(x), max(x), 50)
y_pred = result.predict(pd.DataFrame({"x": xx}))
plt.plot(xx, y_pred)
plt.scatter(x, y)
```

```python
xIvan = [800]
yHatIvan = result.predict(pd.DataFrame({"x": xIvan}))
yHatIvan
```

```python
reg = ols(formula="Y ~ X1 + X2", data=data, )
result = reg.fit()
result.summary()
```
