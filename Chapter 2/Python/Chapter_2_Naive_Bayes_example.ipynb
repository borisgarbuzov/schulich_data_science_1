```
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
import numpy as np
import pprint as pp
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, multilabel_confusion_matrix
from sklearn import datasets



```

```
X = [[1, 1], 
     [2, 2], 
     [3, 1], 
     [1, 3], 
     [2, 1], 
     [3, 1]]
X = np.array(X)
y = np.array([0, 0, 0, 1, 1, 1])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state=42)

clf = MultinomialNB(alpha=1.0, fit_prior=True, class_prior=None)
clf.fit(X=X_train, y=y_train, sample_weight=None)

y_pred = clf.predict(X_test)
print(" prediction - ", y_pred, "\n")
result = confusion_matrix(y_test, y_pred)
print(" Confusion matrix:\n", result, "\n =================\n\n")
result1 = classification_report(y_test, y_pred)
print(" Classification Report:\n", result1, )
```

```
print("coef: ", clf.coef_)
print("intercept: ", clf.intercept_)
print("class_log_prior: ", clf.class_log_prior_)
```

MultinomialNB for iris dataset

```
iris_database = load_iris()
df_iris = pd.DataFrame(iris_database.data, columns=iris_database["feature_names"])
df_iris["target"] = iris_database["target"]
df_iris
```

```
X_train, X_test, y_train, y_test = train_test_split(
   iris_database.data, iris_database["target"], test_size = 0.3, random_state = 1)
```

```
print(" ===== X_train ====\n", X_train[:5], "\n==================")
print("===== y_train ====\n", y_train[:5], "\n==================")
```

```
print(" ===== X_test ====\n", X_test[:5], "\n==================")
print("===== y_test ====\n", y_test[:5], "\n==================")
```

```


clf = MultinomialNB(alpha=1.0, fit_prior=True, class_prior=None)
clf.fit(X=X_train, y=y_train, sample_weight=None)

y_pred = clf.predict(X_test)
print(" prediction - ", y_pred, "\n")

result = confusion_matrix(y_test, y_pred)
print(" Confusion matrix:\n", result, "\n =================\n\n")

result1 = multilabel_confusion_matrix(y_test, y_pred)
print(" multilabel CM:\n", result1, "\n=================\n\n")

result2 = classification_report(y_test, y_pred)
print(" Classification Report:\n", result2, )
```
