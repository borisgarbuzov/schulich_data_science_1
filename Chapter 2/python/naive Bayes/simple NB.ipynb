```python
import numpy as np
rng = np.random.RandomState(1)
X = rng.randint(5, size=(6, 100))
y = np.array([1, 2, 3, 4, 5, 6])


X.shape
```

```python
from sklearn.naive_bayes import CategoricalNB, MultinomialNB
clf = CategoricalNB()
clf.fit(X, y)
```

```python
print(clf.predict(X[2:3]))
```

```python
clf = MultinomialNB()
clf.fit(X, y)
```

```python
print(clf.predict(X[2:3]))
```

```python
X = np.array([[1, 1],
              [2, 2],
              [3, 3],
              [7, 7],
              [5, 5],
              [6, 6]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
clf = MultinomialNB()
clf.fit(X, y)
```

```python
X_test = np.array([[7, 7]])
clf.predict(X_test)
```

```python
clf = CategoricalNB()
clf.fit(X, y)
```

```python
X_test = np.array([[4, 4]])
clf.predict(X_test)
```
