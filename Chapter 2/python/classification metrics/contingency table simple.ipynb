```python
import pandas as pd
import numpy as np
```

```python
myMatrix = np.array([[1, 2], [1, 2], [3, 4], [5, 6]])
df = pd.DataFrame(myMatrix, columns=["x1", "x2"])
df
```

```python
myTable = pd.crosstab(df.x1, df.x2, margins=True)
myTable
```

```python
pd.crosstab(df.x1, df.x2, margins=True, normalize="all")
```

```python
pd.crosstab(df.x1, df.x2, margins=True, normalize="index")
```

```python
pd.crosstab(df.x1, df.x2, margins=True, normalize="columns")
```
