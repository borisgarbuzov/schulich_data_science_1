```python
import numpy as np
x = np.array([[1], 
              [2], 
              [3], 
              [4]])
y = np.array([[1], 
              [3], 
              [5], 
              [4]])
x_train = x[:2]
x_train
```

```python
x_test = x[2:]
x_test
```

```python
y_train = y[:2]
y_test = y[2:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
y_pred_test = reg_model.predict(x_test)
test_mse1 = mean_squared_error(y_test, y_pred_test)
test_mse1
```

Swap test and train

```python
x_train, x_test = x_test, x_train
y_train, y_test = y_test, y_train
```

Now the same code as before, but mse_2

```python
reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
y_pred_test = reg_model.predict(x_test)
test_mse2 = mean_squared_error(y_test, y_pred_test)
test_mse2
```

```python
mse_2fold = np.mean([test_mse1, test_mse2])
mse_2fold
```

Now the same, but using the special library

```python
from sklearn.model_selection import KFold, cross_val_score
kfold = KFold(n_splits = 2)
reg_model = LinearRegression()
mse_array = cross_val_score(estimator=reg_model, X=x, y=y, cv=kfold, scoring='neg_mean_squared_error')
mse_array
```

```python
-np.mean(mse_array)
```

```python
import sklearn
# Look at the possible metrics and find the one corresponding to MSE:
print(sorted(sklearn.metrics.SCORERS.keys()))
```

```python
from sklearn.model_selection import RepeatedKFold
rkf = RepeatedKFold(n_splits=2, n_repeats=2)
reg_model = LinearRegression()
mse_array = cross_val_score(estimator=reg_model, X=x, y=y, cv=rkf, scoring='neg_mean_squared_error')
mse_array, -np.mean(mse_array)
```

Leave one out (LOO). If by hand, we can do 4 fits, take their 4 MSE's and average them up. But it takes space. We do it just by using the lbrary this time. 

```python
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
reg_model = LinearRegression()
mse_array = cross_val_score(estimator=reg_model, X=x, y=y, cv=loo, scoring='neg_mean_squared_error')
mse_array, -np.mean(mse_array)
```
