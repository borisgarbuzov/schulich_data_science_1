```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
X = np.c_[np.random.randint(0, 10, size=10), 
          np.random.randint(0, 10, size=10)]
y = np.random.randint(0, 10, size=10)
```

```python
X = np.array([10, 10.001, 10.002]).reshape(-1, 1)
y = np.array([10, 9.99, 9.98])
```

```python
reg = LinearRegression()
reg.fit(X, y)
reg.coef_, reg.intercept_
```

```python
r2_score(y, reg.predict(X))
```

```python
mean_absolute_error(y, reg.predict(X))
```

```python
valX = [[0.5]]
valY = [0.49]
regPred = reg.predict(valX)
regError = abs(regPred - valY)
regError
```

```python
ridge = Ridge(alpha=0.001)
ridge.fit(X, y)
ridge.coef_, ridge.intercept_
```

```python
r2_score(y, ridge.predict(X))
```

```python
mean_absolute_error(y, ridge.predict(X))
```

```python
ridgePred = ridge.predict(valX)
ridgeError = abs(ridgePred - valY)
ridgeError
```

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X, y)
lasso.coef_, lasso.intercept_
```

```python
r2_score(y, lasso.predict(X))
```

```python
mean_absolute_error(y, lasso.predict(X))
```

```python
lassoPred = lasso.predict(valX)
lassoError = abs(lassoPred - valY)
lassoError
```
