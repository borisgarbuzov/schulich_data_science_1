```python
from sklearn.cross_decomposition import PLSRegression
```

```python
X = [[0., 0., 1.], [1.,0.,0.], [2.,2.,2.], [2.,5.,4.]]
Y = [[0.1, -0.2], [0.9, 1.1], [6.2, 5.9], [11.9, 12.3]]
# Y = [0.1, 0.9, 6.2, 11.9]
```

```python
pls1 = PLSRegression(n_components=1)
pls1.fit(X, Y)

pls2 = PLSRegression(n_components=2)
pls2.fit(X, Y)

pls3 = PLSRegression(n_components=3)
pls3.fit(X, Y)

Y_pred1 = pls1.predict(X)
Y_pred1
```

```python
Y_pred2 = pls2.predict(X)
Y_pred2
```

```python
Y_pred3 = pls3.predict(X)
Y_pred3
```

```python
pls4 = PLSRegression()
pls4.fit(X, Y)
pls4.get_params
```
