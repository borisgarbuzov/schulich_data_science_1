```
import seaborn as sns
%matplotlib inline
```

**Important Features of Seaborn**

Seaborn is built on top of Python’s core visualization library Matplotlib. It is meant to serve as a complement, and not a replacement. However, Seaborn comes with some very important features. Let us see a few of them here. The features help in −

    Built in themes for styling matplotlib graphics
    Visualizing univariate and bivariate data
    Fitting in and visualizing linear regression models
    Plotting statistical time series data
    Seaborn works well with NumPy and Pandas data structures
    It comes with built in themes for styling Matplotlib graphics

In most cases, you will still use Matplotlib for simple plotting. The knowledge of Matplotlib is recommended to tweak Seaborn’s default plots.

```
# list embeded datasets in seaborn
myNames = sns.get_dataset_names()
print("myNames = \n", myNames)
```

```
tips = sns.load_dataset('tips')
tips.head()
```

```
# show distplot with kde
# https://seaborn.pydata.org/generated/seaborn.distplot.html
sns.distplot(tips['total_bill'])
# sns.distplot(tips['total_bill'], kde=False)
```

```
#show data with all bins
sns.distplot(tips['total_bill'], bins=len(tips))
```

```
# join two colums in one plot
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```
# setting plot kind
"""
kind{ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
Kind of plot to draw. See the examples for references to the underlying functions.
"""
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```
# pairplotting any numeric culums
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(tips)
```

```
# categorize plot by data feature (for exumple 'smoker')
sns.pairplot(tips, hue='sex')
```

```
# bar plot to compare data
# https://seaborn.pydata.org/generated/seaborn.barplot.html
sns.barplot(x='sex', y='total_bill', data=tips)
```

```
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

```
# compare counting of group
# https://seaborn.pydata.org/generated/seaborn.countplot.html
sns.countplot(x='day', data=tips)
```

```
# box_plot
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```
# violon plot
# https://seaborn.pydata.org/generated/seaborn.violinplot.html
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```
# show all observations along with some 
# representation of the underlying distribution
# https://seaborn.pydata.org/generated/seaborn.stripplot.html
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```
# flights dataset
fly = sns.load_dataset('flights')
fc = fly.corr()
```

```
fly.head()
```

```
# heatmap
sns.heatmap(fc)
```

```
tc = tips.corr()
sns.heatmap(tc, annot=True)
```

```
# heatmap for flight action
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
fvt = fly.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(fvt, linewidths=0)
fvt
```

```
# clastering
# https://seaborn.pydata.org/generated/seaborn.clustermap.html
sns.clustermap(fvt)
```

```
# regression
# https://seaborn.pydata.org/generated/seaborn.lmplot.html
tips.head()
```

```
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue='sex')
```

```
# regplot with day
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue='sex', col='day', row='time')
```
