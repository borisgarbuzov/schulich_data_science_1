```python
X = np.array([1, 2, 3, 6, 7]).reshape(-1, 1)
y = np.array([1, 1, 1, 2, 2])
```

```python
lda = LinearDiscriminantAnalysis()
lda.store_covariance = True
lda.fit(X, y)
```
